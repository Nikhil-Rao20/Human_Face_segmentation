```python
import os
import shutil

# Define input and output folders
input_folder = 'clean_images'
output_folder = 'Final_Images'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Initialize a counter for renaming
counter = 0

# Iterate through the files in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is a valid image file (you can add more extensions as needed)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        # Construct the new filename with a sequential number
        new_filename = f"{counter:04d}.jpg"  # You can change the extension as needed
        counter += 1

        # Get the full path of the input and output files
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, new_filename)

        # Copy the file from the input folder to the output folder with the new name
        shutil.copy(input_path, output_path)

print("Images renamed and saved in the 'renamed_images' folder.")
```
